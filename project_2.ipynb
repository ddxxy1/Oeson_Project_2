# ***Insights from this project***

In analyzing the dataset from the famous job portal in 2023, several insights were derived:



1.  **Rapid Growth in Job Opportunities:** Statistical evidence indicates a rapid increase in the number of jobs each year, with a staggering 4.53-fold increase between 2022 and 2023.
2.  **Significant Increase in Average Salary:** The average salary in 2023 saw a remarkable 1.457-fold increase compared to the total in 2021, representing a substantial 45.7% surge. This suggests an increasing attractiveness of positions in the data field, possibly due to skill demands and heightened market competition.

3.  **Job Distribution Chart:** Data Engineer, Data Scientist, Data Analyst, and Machine Learning Engineer dominate job positions, predicting intensified competition for these roles in the future.

4.  **Salary Disparities Among Positions:** A significant direct pay gap is observed between BI Developers and Research Scientists, likely stemming from differences in the complexity of their job content.

5.  **Salary Ranking Across Positions:** Applied Scientists claim the highest salaries among common roles, followed closely by Machine Learning Engineers. Subsequently, salaries decline by around 20%, with Data Scientists ranking third, Data Engineers fourth, and Data Analysts fifth.

6.  **Job Category Distribution Chart:** Data Science and Research dominate, followed by Data Engineering, Data Analysis, and Machine Learning and AI. Aligning this with the job category distribution, it's evident that the dominating categories also command higher salaries.

7.  **Impact of Geographic Location on Salaries:** Examination of the top 10 average salaries based on employee and company location reveals that companies in Qatar offer the highest salaries, reaching up to $300,000 annually. Other regions provide similar salaries ranging between $160,000 and $200,000 annually.

8.  **Effect of Experience Level on Salaries:** The average salaries by experience level chart indicates that salaries increase with work experience, with the most substantial increase occurring from entry-level to senior level.

9.  **Influence of Employment Type, Work Setting, and Company Size on Salaries:** Contractors and full-time employees have the highest salaries, with in-person and remote work styles yielding the most pay. Regarding company size, medium and large companies offer the best salaries, with small companies providing nearly 65% of the salaries offered by medium-sized companies.

**Frenquency of each year**

Statistically, the number of jobs is going up every year. Shockingly, the number of jobs has increased 4.53 times between the years 2022 and 2023.

```python
work_year_counts = test['work_year'].value_counts()

print(work_year_counts)
```

Calculated the % increase from year 2022 to 2023

```python
jobs_2023 = test[test['work_year'] == 2023].shape[0]
jobs_2022 = test[test['work_year'] == 2022].shape[0]
jobs_difference = jobs_2023 / jobs_2022
print(jobs_difference)
```

```python
plt.figure(figsize=(10, 6))
work_year_distribution = sns.countplot(x='work_year', data=test, palette='viridis')
plt.title('Frequency of Work Years')
plt.xlabel('Work Year')
plt.ylabel('Count')

work_year_distribution.set_xticklabels(work_year_distribution.get_xticklabels(), rotation=45, horizontalalignment='right')

for bar in work_year_distribution.patches:
    plt.annotate(f'{bar.get_height()}',
                 (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                 ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```

**Average Salary Per Year**

The average salary in 2023 is 1.457 times higher compared to the 2021 total, or a 45.7% increase in average salary.

```python
average_salaries_by_year = test.groupby('work_year')['salary_in_usd'].mean()

# Step 3: Plot the average salaries by work year
plt.figure(figsize=(10, 6))
salary_by_year_plot = sns.barplot(x=average_salaries_by_year.index, y=average_salaries_by_year, palette='viridis')
plt.title('Average Salaries by Work Year')
plt.xlabel('Work Year')
plt.ylabel('Average Salary (USD)')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Add annotations for each salary
for i, value in enumerate(average_salaries_by_year):
    plt.text(i, value + 100, f'${value:.2f}', ha='center', va='bottom', color='black', fontsize=8)

# Show plot
plt.show()
```

% of salary increased

```python
average_salary_2022 = test[test['work_year'] == 2022]['salary_in_usd'].mean()
average_salary_2021 = test[test['work_year'] == 2021]['salary_in_usd'].mean()
average_salary_2023 = test[test['work_year'] == 2023]['salary_in_usd'].mean()

salary2021_2022difference = average_salary_2022 / average_salary_2021
salary2022_2023difference = average_salary_2023 / average_salary_2022

salary_multiplier_2023_vs_2021 = salary2021_2022difference * salary2022_2023difference
print("salary_multiplier_2023_vs_2021:", salary_multiplier_2023_vs_2021)
```

**Highest Salary VS. Lowest Salary**

As we can see from the comparison chart below, the direct pay gap between BI developer and Research scientist is very large. The reason for this large gap is most likely due to the difference in the complexity of their work content.

```python
plt.figure(figsize=(8, 6))
colors = ["green", "red"]
bars = plt.bar(["Top Highest Earner", "Bottom Lowest Earner"],
               [highest_salary, lowest_salary], color=colors)

plt.xticks([0, 1], [test[test['salary_in_usd'] == highest_salary]['job_title'].values[0],
                   test[test['salary_in_usd'] == lowest_salary]['job_title'].values[0]])

plt.ylabel("Salary (USD)", size=12)
plt.title("Highest-Paying vs. Lowest-Paying Job Titles", size=14)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

for bar, salary in zip(bars, [highest_salary, lowest_salary]):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 500,
             f"${salary:.2f}", ha='center', va='bottom', color='black')

plt.show()
```

**Average Salary for the 5 most common job title**

Applied scientists have the highest salaries of the most common roles, with a very small difference in second place for machine learning engineers, after which salaries start to fall by roughly 20 percent, with data scientists in third place, data engineers in fourth place, and data analysts in fifth place.

```python
top_5_jobs = test['job_title'].value_counts().head(5).index
filtered_test = test[test['job_title'].isin(top_5_jobs)]

average_salary_by_job = filtered_test.groupby('job_title')['salary_in_usd'].mean().reset_index()
```

```python
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='salary_in_usd', y='job_title', data=average_salary_by_job, palette='viridis')
barplot.set_xlim(50000, max(average_salary_by_job['salary_in_usd']))
plt.title('Average Salary for Top 5 Most Common Job Titles')
plt.xlabel('Average Salary (USD)')
plt.ylabel('Job Title')

for index, value in enumerate(average_salary_by_job['salary_in_usd']):
    barplot.text(value, index, f'${value:.2f}', ha='left', va='center', fontsize=10, color='black')

plt.show()
```

**Distribution of Job Categories**

The job category map shows that data science and research dominates, followed by data engineering, data analysis and machine learning and AI.

```python
plt.figure(figsize=(12, 8))
job_category_distribution = sns.countplot(x='job_category', data=test, palette='viridis', order=test['job_category'].value_counts().index)
plt.title('Distribution of Job Categories')
plt.xlabel('Job Category')
plt.ylabel('Count')

job_category_distribution.set_xticklabels(job_category_distribution.get_xticklabels(), rotation=45, horizontalalignment='right')

for bar, count in zip(job_category_distribution.patches, test['job_category'].value_counts().values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, f'{count}', ha='center', va='bottom', color='black')

plt.show()
```

**Average Salaries by Job Category**

Pairing the chart below with the job category distribution chart, we can see that the job categories that dominate are also the ones that command the highest salaries. machine learning and AI are among the highest paid.

```python
average_salaries_jc = test.groupby('job_category')['salary_in_usd'].mean().sort_values()

plt.figure(figsize=(12, 8))
average_salary_by_category = sns.barplot(x='job_category', y='salary_in_usd', data=test, order=average_salaries_jc.index, palette='viridis')
plt.title('Average Salaries by Job Category (Ascending Order)')
plt.xlabel('Job Category')
plt.ylabel('Average Salary (USD)')

plt.xticks(rotation=45, ha='right')

for bar, salary in zip(average_salary_by_category.patches, test.groupby('job_category')['salary_in_usd'].mean().values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100, f'${salary:.2f}', ha='center', va='bottom', color='black')

plt.show()
```

**Aevrage Salaries in USD by Currency**

```python
average_salaries_by_currency = test.groupby('salary_currency')['salary_in_usd'].mean().sort_values()

plt.figure(figsize=(12, 8))
average_salary_by_currency_plot = sns.barplot(x='salary_currency', y='salary_in_usd', data=test, order=average_salaries_by_currency.index, palette='viridis')
plt.title('Average Salaries in USD by Currency')
plt.xlabel('Currency')
plt.ylabel('Average Salary (USD)')

plt.xticks(rotation=45, ha='right')

for bar, salary, color in zip(average_salary_by_currency_plot.patches, average_salaries_by_currency.values, sns.color_palette('viridis')):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100, f'${salary:.2f}', ha='center', va='bottom', color=color)

plt.show()
```

**Top 10 Average Salaries by Employee Residence  and Company Location**

Looking at the list of the top 10 average salaries by employee's place of residence and company's location, we can observe that companies in the region of Qatar have the highest salaries, which are up to 300,000USD per year, while the rest of the nine regions offer similar salaries, which are between 160,000USD and 200,000USD per year.

```python
top_10_salaries_by_location = (
    test.groupby(['employee_residence', 'company_location'])['salary_in_usd']
    .mean()
    .nlargest(10)
    .unstack()
)

plt.figure(figsize=(16, 8))
top_10_salary_by_location_plot = top_10_salaries_by_location.plot(kind='bar', stacked=False, cmap='viridis', width=.7)
plt.title('Top 10 Average Salaries by Employee Residence and Company Location')
plt.xlabel('Location')
plt.ylabel('Average Salary (USD)')
plt.legend(title='Company Location', bbox_to_anchor=(1, 1))

plt.xticks(rotation=45, ha='right')

plt.ylim(80000, plt.ylim()[1])

for i, (index, row) in enumerate(top_10_salaries_by_location.iterrows()):
    for col, value in row.items():
        if np.isfinite(value):
            plt.text(i, value + 100, f'${value:.2f}', ha='left', va='bottom', color='black', fontsize=8)

plt.show()
```

**Average Salaries by Experience Level**

Looking at the picture of average salaries by experience level, we can conclude that salaries do increase with work experience, and that the largest increase in salaries occurs during the period from entry-level to senior level.

```python
average_salaries_by_experience = test.groupby('experience_level')['salary_in_usd'].mean()

experience_levels_order = ['Entry-level', 'Mid-level', 'Senior', 'Executive']

average_salaries_by_experience = average_salaries_by_experience.reindex(experience_levels_order)

plt.figure(figsize=(10, 6))
salary_by_experience_plot = sns.barplot(x=average_salaries_by_experience.index, y=average_salaries_by_experience, palette='viridis')
plt.title('Average Salaries by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Average Salary (USD)')

plt.xticks(rotation=45, ha='right')

for i, value in enumerate(average_salaries_by_experience):
    plt.text(i, value + 100, f'${value:.2f}', ha='center', va='bottom', color='black', fontsize=8)

plt.show()
```

**Visualize average salary based on employment type, work setting and company_size**

Combined with the three bar charts below, we can conclude that contractors and full-time employees have the highest salaries, with in-person and remote work styles having the most pay. Then in terms of company size medium and large companies offer the best salaries, with small companies offering almost 65 per cent of the salaries offered by medium-sized companies.
